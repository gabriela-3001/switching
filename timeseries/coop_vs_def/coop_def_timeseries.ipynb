```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from collections import Counter
from matplotlib.colors import ListedColormap, BoundaryNorm, Normalize
```

```python
df_50 = pd.read_csv('coop_def/suc_50/5101', sep = '\s+', header = None)
columns = ['time', 'sucrose', 'glucose', 'cooperator', 'defector']
df_50.columns = columns
```

```python
df_100 = pd.read_csv('coop_def/suc_100/5101', sep = '\s+', header = None)
columns = ['time', 'sucrose', 'glucose', 'cooperator', 'defector']
df_100.columns = columns
```

```python
df_90 = pd.read_csv('coop_def/suc_90/5101', sep = '\s+', header = None)
columns = ['time', 'sucrose', 'glucose', 'cooperator', 'defector']
df_90.columns = columns
```

```python
types = ['cooperator', 'defector']
colors = ['blue', 'red']
```

```python
fig, ax = plt.subplots(2,2, figsize = (16,12))

ax[0,0].set_ylabel('Frequencies', fontsize = 16)  # Set the label for the y-axis

for j in range(2):
    ax[0,0].plot(df_90['time'], df_90[types[j]] / df_90[types].sum(axis = 1), color=colors[j], linewidth = 3)
    ax[0,1].plot(df_50['time'], df_50[types[j]] / df_50[types].sum(axis = 1), color=colors[j], linewidth = 3)

ax[0,0].plot(df_90['time'],df_90['sucrose'], color='k', linewidth = 2)
ax[0,0].plot(df_90['time'],df_90['glucose'], color='silver', linewidth = 2)
#ax2.legend(['Sucrose', 'Glucose'], loc = 'upper right', fontsize = 16)
ax[0,0].set_ylabel('Frequencies, Sugar levels', fontsize = 16)  # Set the label for the y-axis

ax[0,1].plot(df_50['time'],df_50['sucrose'], color='k', linewidth = 2)
ax[0,1].plot(df_50['time'],df_50['glucose'], color='silver', linewidth = 2)
#ax3.legend(['Sucrose', 'Glucose'], loc = 'upper right', fontsize = 16)
#ax[0,1].set_ylabel('Frequencies, Sugar levels', fontsize = 16)  # Set the label for the y-axis

for j in range(2):
    ax[1,0].plot(df_90['time'],df_90[types[j]], color = colors[j], linewidth = 3)
    ax[1,1].plot(df_50['time'],df_50[types[j]], color = colors[j], linewidth = 3)

ax[1,0].set_ylabel('Abundances', fontsize = 16)

#ax2.tick_params(labelsize = 15)
#ax3.tick_params(labelsize = 15)

ax[0,1].set_title('Sucrose 50%', fontsize = 30)
ax[0,0].set_title('Sucrose 90%', fontsize = 30)

for i in range(2):
    ax[1,i].set_xlabel('Time', fontsize = 16)
    ax[0,i].set_ylim(-0.05,1.05)
    ax[1,i].set_ylim(-0.05,3)
    for j in range(2):
        ax[i,j].tick_params(labelsize = 15)
        ax[i,j].set_xlim(3000,5000)

plt.tight_layout()
plt.savefig('timeseries_coop_def_JUL12.png', dpi = 600, bbox_inches = 'tight')
```
