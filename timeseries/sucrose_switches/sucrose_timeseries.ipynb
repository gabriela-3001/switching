```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from collections import Counter
from matplotlib.colors import ListedColormap, BoundaryNorm, Normalize
```

```python
df_50 = pd.read_csv('sucrose/suc_50/6101', sep = '\s+', header = None)
columns = ['time', 'sucrose', 'glucose', 'cooperator', '0.25', '0.5', '0.75', 'defector']
df_50.columns = columns
```

```python
df_100 = pd.read_csv('sucrose/suc_100/6101', sep = '\s+', header = None)
columns =  ['time', 'sucrose', 'glucose', 'cooperator', '0.25', '0.5', '0.75', 'defector']
df_100.columns = columns
```

```python
colors = ['blue', 'green', 'yellow', 'orange', 'red']
types = ['cooperator', '0.25', '0.5', '0.75', 'defector']
```

```python
fig, ax = plt.subplots(2,2, figsize = (16,12))

ax[0,0].set_ylabel('Frequencies', fontsize = 16)  # Set the label for the y-axis

for j in range(5):
    ax[0,0].plot(df_100['time'], df_100[types[j]] / df_100[types].sum(axis = 1), color=colors[j], linewidth = 3)
    ax[0,1].plot(df_50['time'], df_50[types[j]] / df_50[types].sum(axis = 1), color=colors[j], linewidth = 3)


ax[0,0].plot(df_100['time'],df_100['sucrose'], color='k', linewidth = 2)
ax[0,0].plot(df_100['time'],df_100['glucose'], color='silver', linewidth = 2)
#ax[1,0].legend(['Sucrose', 'Glucose'], loc = 'upper right', fontsize = 16)
ax[0,0].set_ylabel('Frequencies, Sugar levels', fontsize = 16)  # Set the label for the y-axis

ax[0,1].plot(df_50['time'],df_50['sucrose'], color='k', linewidth = 2)
ax[0,1].plot(df_50['time'],df_50['glucose'], color='silver', linewidth = 2)
#ax[0,1].legend(['Sucrose', 'Glucose'], loc = 'upper right', fontsize = 16)
#ax[0,1].set_ylabel('Frequencies, Sugar levels', fontsize = 16)  # Set the label for the y-axis

for j in range(5):
    ax[1,0].plot(df_100['time'],df_100[types[j]], color = colors[j], linewidth = 3)
    ax[1,1].plot(df_50['time'],df_50[types[j]], color = colors[j], linewidth = 3)

ax[1,0].set_ylabel('Abundances', fontsize = 16)

ax[0,1].set_title('Sucrose 50%', fontsize = 30)
ax[0,0].set_title('Sucrose 100%', fontsize = 30)

for i in range(2):
    ax[1,i].set_xlabel('Time', fontsize = 16)
    ax[0,i].set_ylim(-0.05,1.05)
    ax[1,i].set_ylim(-0.05,2.5)
    for j in range(2):
        ax[i,j].tick_params(labelsize = 15)
        ax[i,j].set_xlim(500,1500)

plt.tight_layout()
plt.savefig('timeseries_sucrose_JUL12.png', dpi = 600, bbox_inches = 'tight')
```

```python
plt.plot(df_100['time'],df_100['sucrose'], color='k', linewidth = 2)
plt.plot(df_100['time'],df_100['glucose'], color='silver', linewidth = 2)

for j in range(5):
    plt.plot(df_100['time'], df_100[types[j]] / df_100[types].sum(axis = 1), color=colors[j], linewidth = 3)
    
plt.legend(['Sucrose', 'Glucose', "Cooperator", 'C > 0.25', 'C > 0.5', 'C > 0.75','Defector'], fontsize = 14, bbox_to_anchor = (2,1.2), ncol = 7)
plt.savefig('timeseries_sucrose_legend_JUL12.png', dpi = 600, bbox_inches = 'tight')
```
