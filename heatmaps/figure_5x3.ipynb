```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from collections import Counter
from matplotlib.colors import ListedColormap, BoundaryNorm, Normalize
```

```python
from itertools import product

c_range = list(np.arange(0,1.01,0.01))
s_range = list(np.arange(0,1.01,0.01))

combs = list(product(s_range, c_range))
comb_pos = list(product(list(np.arange(0,101,1)), list(np.arange(0,101,1))))
dict_pos = dict(zip(combs, range(1, 10202)))
dict_pos_invert = dict(zip(range(1, 10202), combs))
dict_pos_invert_pos = dict(zip(range(1, 10202), comb_pos))
```

```python
colors = ['blue', 'green', 'yellow', 'orange', 'red']
directories_fitness = ['fitness/suc_100/combined.txt',
              'fitness/suc_90/combined.txt', 
              'fitness/suc_50/combined.txt',
              'fitness/suc_10/combined.txt',
              'fitness/suc_0/combined.txt']

directories_sucrose = ['sucrose/suc_100/combined.txt',
              'sucrose/suc_90/combined.txt', 
              'sucrose/suc_50/combined.txt',
              'sucrose/suc_10/combined.txt',
              'sucrose/suc_0/combined.txt']

directories_glucose = ['glucose/suc_100/combined.txt',
              'glucose/suc_90/combined.txt', 
              'glucose/suc_50/combined.txt',
              'glucose/suc_10/combined.txt',
              'glucose/suc_0/combined.txt']

directories_CD = ['coop_def/suc_100/combined.txt',
              'coop_def/suc_90/combined.txt', 
              'coop_def/suc_50/combined.txt',
              'coop_def/suc_10/combined.txt',
              'coop_def/suc_0/combined.txt']

all_directories = [directories_fitness, directories_glucose, directories_sucrose, directories_CD]
all_directories = [directories_CD, directories_sucrose, directories_glucose, directories_fitness]
names_dirs = ['CD','sucrose', 'glucose','fitness']
```

```python
dict_all = {}

for S in range(len(all_directories)):
    
    print(names_dirs[S])

    dict_res = {}

    for D in range(5):  

        data_sim = pd.read_table(all_directories[S][D], sep = '\s+', header = None)
        
        if names_dirs[S] == 'fitness':
            switches = ['1', '2', '3']
            data_sim.columns = switches

            data_sim['m'] = list(range(1, 10202))
            data_sim['i'] = [dict_pos_invert_pos[x][1] for x in range(1,10202)]
            data_sim['j'] = [dict_pos_invert_pos[x][0] for x in range(1,10202)]

            data_sim['s'] = [s_range[dict_pos_invert_pos[x][1]] for x in range(1,10202)]
            data_sim['c'] = [c_range[dict_pos_invert_pos[x][0]] for x in range(1,10202)]

            data_sim['max_value'] = data_sim[['1', '2', '3']].max(axis=1)
            #data_sim['column_with_max'] = data_sim[['1', '2', '3']].idxmax(axis=1)
            data_sim['column_with_max'] = data_sim[['1', '2', '3']].apply(lambda row: np.random.choice(row[row == row.max()].index), axis=1)

            # Calculate the second highest value by sorting and picking the second last item
            data_sim['second_highest'] = data_sim[['1', '2', '3']].apply(lambda x: x.sort_values(ascending=False).iloc[1], axis=1)

            data_sim['ratio_max_to_second'] = data_sim['max_value'] / data_sim['second_highest']
            data_sim['second_highest_index'] = data_sim[switches].apply(lambda x: x.drop(x.idxmax()).idxmax(), axis=1)
            
            data_sim['total_abundance'] = data_sim[switches].sum(axis = 1)
            
            data_sim['cooperator'] = data_sim['1']

            dict_res[D] = data_sim
            
        elif names_dirs[S] == 'CD':
            switches = ['1', '2']
            data_sim.columns = switches

            data_sim['m'] = list(range(1, 10202))
            data_sim['i'] = [dict_pos_invert_pos[x][1] for x in range(1,10202)]
            data_sim['j'] = [dict_pos_invert_pos[x][0] for x in range(1,10202)]

            data_sim['s'] = [s_range[dict_pos_invert_pos[x][1]] for x in range(1,10202)]
            data_sim['c'] = [c_range[len(c_range)-1-dict_pos_invert_pos[x][0]] for x in range(1,10202)]

            data_sim['max_value'] = data_sim[switches].max(axis=1)
            #data_sim['column_with_max'] = data_sim[switches].idxmax(axis=1)
            data_sim['column_with_max'] = data_sim[['1', '2']].apply(lambda row: np.random.choice(row[row == row.max()].index), axis=1)

            # Calculate the second highest value by sorting and picking the second last item
            data_sim['second_highest'] = data_sim[switches].apply(lambda x: x.sort_values(ascending=False).iloc[1], axis=1)

            data_sim['ratio_max_to_second'] = data_sim['max_value'] / data_sim['second_highest']
            data_sim['second_highest_index'] = data_sim[switches].apply(lambda x: x.drop(x.idxmax()).idxmax(), axis=1)
            
            data_sim['cooperator'] = data_sim['1']
            data_sim['total_abundance'] = data_sim[switches].sum(axis = 1)

            dict_res[D] = data_sim
            
        else:
            switches = ['1', '2', '3', '4', '5']
            data_sim.columns = switches

            data_sim['m'] = list(range(1, 10202))
            data_sim['i'] = [dict_pos_invert_pos[x][1] for x in range(1,10202)]
            data_sim['j'] = [dict_pos_invert_pos[x][0] for x in range(1,10202)]

            data_sim['s'] = [s_range[dict_pos_invert_pos[x][1]] for x in range(1,10202)]
            data_sim['c'] = [c_range[len(c_range)-1-dict_pos_invert_pos[x][0]] for x in range(1,10202)]

            data_sim['max_value'] = data_sim[switches].max(axis=1)
            #data_sim['column_with_max'] = data_sim[switches].idxmax(axis=1)
            data_sim['column_with_max'] = data_sim[['1', '2', '3', '4', '5']].apply(lambda row: np.random.choice(row[row == row.max()].index), axis=1)

            # Calculate the second highest value by sorting and picking the second last item
            data_sim['second_highest'] = data_sim[switches].apply(lambda x: x.sort_values(ascending=False).iloc[1], axis=1)
            data_sim['second_highest_index'] = data_sim[switches].apply(lambda x: x.drop(x.idxmax()).idxmax(), axis=1)

            data_sim['ratio_max_to_second'] = data_sim['max_value'] / data_sim['second_highest']
            
            data_sim['total_abundance'] = data_sim[switches].sum(axis = 1)
            data_sim['cooperator'] = data_sim['1']

            dict_res[D] = data_sim           
        
    dict_all[names_dirs[S]] = dict_res
```

```python
all_mat = {}

for name in names_dirs:
    
    print(name)

    dict_mat = {}

    for D in range(5):    

        res_matrix = np.zeros((101,101))

        for i in range(101):
            for j in range(101):

                df_res = dict_all[name][D][(dict_all[name][D]['i'] == i) & (dict_all[name][D]['j'] == j)]

                if float(df_res['max_value']) > 0.01:

                    res_matrix[100-i,j] = int(df_res['column_with_max'])

        dict_mat[D] = res_matrix
        
    all_mat[name] = dict_mat
```
